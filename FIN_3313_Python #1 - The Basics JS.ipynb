```python
#@title # 5. Display the female gold medal winners for the 100m track and field sprint event over the years.

print('Female Gold Medal Winners for 100m Track and Field Sprint Event Over the Years')
print('-------------------------------------------------------------------------------')
oo[(oo.Medal == 'Gold') & (oo.Gender == 'Women') & (oo.Event == '100m')]
```

```python
#@title # 6. Plot the number of medals achieved by the U.S. team (men and women) in Beijing 2008 using both Matplotlib and Seaborn.

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Number of Medals Achieved by the U.S. Team (Men and Women) in Beijing 2008')
print('--------------------------------------------------------------------')
mw = oo[(oo.Edition == 2008) & (oo.NOC == 'USA')]
mw.head()

print('MatplotLib Plot')
mw.Gender.value_counts().plot(kind='bar');
plt.show()


print('')

print('Seaborn Plot')
sns.countplot(data =mw , x='Gender');
plt.show()


```

```python
#@title # 9. Using groupby(), plot the total number of medals awarded at each of the Olympic games throughout history.
print('Total Medals Awarded at Each Olympic Game from 1896-2008')
print('----------------------------------------')
oo.groupby('Edition').size().plot()
```
